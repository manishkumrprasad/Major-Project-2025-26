```python
import matplotlib.pyplot as plt

x = [2 , 4 , 6 , 8 , 10 ]
y = [ 1 , 2 , 3 , 6 , 9 ]
c = ["r" , "g" , "y" ,  "b" , "r"] #boring way to add colors

co_map = [34  , 18 , 78 , 67 , 56]


sizes = [30 , 45 , 100 , 150  , 250 ] 
#plt.scatter(x , y , color = c , s = sizes , alpha = 0.5 , visible = True , marker = "*" , linewidth = 5 , cmap = "viridis") #through alpha parameter we can change the opacity of the dot 

plt.scatter(x , y , s = sizes , c = co_map , cmap = "viridis") #through alpha parameter we can change the opacity of the dot 
plt.colorbar()
plt.title("Scatter Plot" , fontsize = 20 )
plt.xlabel("X-Axis" , fontsize = 15 )
plt.ylabel("Y-Axis")
plt.show()
```
