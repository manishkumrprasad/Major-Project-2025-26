```python
import matplotlib.pyplot as plt

x = ["Haryana" , "Himachal Pradesh" , "Kerala" , "Nagaland" , "Jammu Kashmir" , "Punjab" ]
y = [ 10 , 50 ,30 ,20 ,50 ,60]
z = [ 8 , 40 , 12 , 12 , 12 , 32]
c = ["r" , "g" , "y" , "b" , "g" , "y"]

plt.xlabel("States" )
plt.ylabel("GDP In Crore")
plt.title("GDP Ranking Of Different States Of India", fontsize = 20)

plt.bar(x , y , width = 0.2 , align  ="center" , edgecolor = "b" ,color = "r" , label = "GDP") #Showing the current year data
plt.bar(x , z , width = 0.4 , align  ="center" , edgecolor = "b" ,color = "g" , label = "GDP") #Showing the previous year data

#The red line shows this is gdp data
#And The Green Lines Shows The Previous year data

plt.legend()
plt.show()
```
